```python
import json

with open(f'statistics/avaibility_tests.json', 'r') as file:
    data = json.load(file)
```

```python
list(data.keys())[0]
```

```python
list(data.values())[0][0]['initial_cost']
```

```python
pandas_data = {'probabilities (lecturer, room)': [], 'runtime': [], 'cost function change': []}
for key, values in data.items():
    for value in values:
        pandas_data['probabilities (lecturer, room)'].append(key)
        pandas_data['runtime'].append(value['elapsed_time'])
        pandas_data['cost function change'].append(value['initial_cost'] - value['best_cost'])
```

```python
import pandas as pd

df = pd.DataFrame(pandas_data)
df.head()
```

```python
df['probabilities (lecturer, room)'].value_counts()
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.boxplot(x='probabilities (lecturer, room)', y='cost function change', data=df, showfliers=False).set_title(
    "Wykres prawdopodobieństw początkowego zapełnienia macierzy dostępności prowadzącego i sali od wartości funkcji kosztu")
```
