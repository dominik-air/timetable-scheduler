```python
import timetable_scheduler

@timetable_scheduler.timing
def test_instance_creator(term_id: int, **kwargs):
    results = {}
    iterations = 0
    for lecturer_p, room_p in zip([0.5, 0.7, 0.6, 0.4, 0.7, 0.3, 1], [0.5, 0.7, 0.4, 0.6, 0.3, 0.7, 1]):
        timetable_scheduler.create_dataset(term_id=term_id, lecturer_p=lecturer_p, room_p=room_p)
        results[(lecturer_p, room_p)] = []
        setup = timetable_scheduler.StatisticalTestsAlgorithmSetup(
            cooling_schedule=timetable_scheduler.simulated_annealing.exponential_cooling_schedule,
            operator_probabilities=[0.1, 0.1, 0.8],
            Tmax=20, Tmin=2.5, kmax=5, alpha=0.99, **kwargs)
        for i in range(30):
            result = setup.SA()
            results[(lecturer_p, room_p)].append(result)
            iterations += 1
            print('-'*10, iterations, '-'*10)
    return results
```

```python
results, time = test_instance_creator(term_id=5)
```

```python
json_data = {str(parameters): [r.to_json() for r in result] for parameters, result in results.items()}
```

```python
import json
with open('statistics/avaibility_tests.json', 'w') as f:
    json.dump(json_data, f, indent=4)
```
