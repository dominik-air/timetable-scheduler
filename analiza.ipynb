```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

with open('statistics/init_solutions.txt', 'r') as f:
    data = [float(i) for i in f.readline().split(',')]

df = pd.DataFrame(data)
sns.histplot(data=df)
plt.xlabel('wartość funkcji celu')
plt.ylabel('ilość wystąpień')
plt.title('Rozkład wartości funkcji celu rozwiązania początkowego')
plt.legend()
plt.show()
```
