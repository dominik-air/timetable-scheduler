```python
import timetable_scheduler

@timetable_scheduler.timing
def test_instance_creator(term_id: int, availability_p: float, **kwargs):
    results = {}
    iterations = 0
    timetable_scheduler.create_dataset(term_id=term_id, lecturer_p=availability_p, room_p=availability_p)
    for matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation in zip([0.1, 0.15, 0.2, 0.25, 0.3], [0.1, 0.2, 0.1, 0.3, 0.15], [0.8, 0.65, 0.7, 0.45, 0.55]):
        results[(matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)] = []
        setup = timetable_scheduler.StatisticalTestsAlgorithmSetup(
            cooling_schedule=timetable_scheduler.simulated_annealing.exponential_cooling_schedule,
            operator_probabilities=[matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation],
            Tmax=20, Tmin=2.5, kmax=5, alpha=0.99, **kwargs)
        for i in range(30):
            result = setup.SA()
            results[(matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)].append(result)
            iterations += 1
            print('-'*10, iterations, '-'*10)
    return results
```

```python
results, time = test_instance_creator(term_id=5, availability_p=0.7)
```

```python
json_data = {str(parameters): [r.to_json() for r in result] for parameters, result in results.items()}
```

```python
import json
with open('statistics/operators_tests.json', 'w') as f:
    json.dump(json_data, f, indent=4)
```
