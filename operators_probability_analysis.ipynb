```python
import json

with open(f'statistics/operators_tests.json', 'r') as file:
    data = json.load(file)
```

```python
list(data.keys())[0]
```

```python
list(data.values())[0][0]['initial_cost']
```

```python
pandas_data = {'probabilities (matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)': [], 'runtime': [], 'cost function change': []}
for key, values in data.items():
    for value in values:
        pandas_data['probabilities (matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)'].append(key)
        pandas_data['runtime'].append(value['elapsed_time'])
        pandas_data['cost function change'].append(value['initial_cost'] - value['best_cost'])
```

```python
import pandas as pd

df = pd.DataFrame(pandas_data)
df.head()
```

```python
df['probabilities (matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)'].value_counts()
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(15.7,8.27)})
sns.boxplot(x='probabilities (matrix_transposition, matrix_inner_translation, matrix_cut_and_paste_translation)', y='cost function change', data=df, showfliers=False).set_title(
    "Wykres prawdopodobieństw operatorów generowania sąsiedztwa od wartości funkcji kosztu")
```
